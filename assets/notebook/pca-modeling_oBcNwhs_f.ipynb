### Importing the Features and target

```python
import os, pandas as pd
features = pd.read_csv('/project_data/data_asset/features.csv')
```

```python
import os, pandas as pd
target = pd.read_csv('/project_data/data_asset/target.csv')
```

```python
X = features.iloc[:,1:].values
X[0:2]
```

```python
y = target['0'].values
y[0:5]
```

### Splitting the X and y into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

### Fitting Logistic Regression to the Training set

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

### Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
y_pred[0:5]
```

### Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

### Saving the Model

```python


from watson_machine_learning_client import WatsonMachineLearningAPIClient
import os

token = os.environ['USER_ACCESS_TOKEN']

wml_credentials = {
   "token": token,
   "instance_id" : "openshift",
   "url": os.environ['RUNTIME_ENV_APSX_URL'],
   "version": "2.5.0"
}

client = WatsonMachineLearningAPIClient(wml_credentials)
```

```python
meta_props={
   client.repository.ModelMetaNames.NAME: "PCAModel",
   client.repository.ModelMetaNames.RUNTIME_UID: "scikit-learn_0.20-py3.6",
   client.repository.ModelMetaNames.TYPE: "scikit-learn_0.20"     
}

project_id = os.environ['PROJECT_ID']
client.set.default_project(project_id)

```

```python
 model_artifact = client.repository.store_model(classifier,
                                             meta_props=meta_props,
                                             training_data=X_train,
                                             training_target=y_train)
 model_uid = client.repository.get_model_uid(model_artifact)
 print("Model UID = " + model_uid)
```
